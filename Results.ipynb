# Simple CNN: PGD Vs FGSM

```python
import matplotlib.pyplot as plt

# Simple CNN - PGD
simple_cnn_accuracy = [97.19, 99.15, 99.44, 99.59, 99.67]
simple_cnn_loss = [175.4103, 53.0943, 34.6272, 24.3543, 20.9508]

# Simple CNN - FGSM
simple_cnn_loss_fgsm = [0.0958, 0.0289, 0.0186, 0.0133, 0.0120]
simple_cnn_accuracy_fgsm = [97.02, 99.11, 99.43, 99.58, 99.65]


epochs = range(1, len(simple_cnn_accuracy) + 1)

# Accuracy vs Epoch
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs, simple_cnn_accuracy, label="Simple CNN - PGD", marker='o')
plt.plot(epochs, simple_cnn_accuracy_fgsm, label="Simple CNN - FGSM", marker='s')
plt.title("Accuracy vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.grid(True)

# Loss vs Epoch
plt.subplot(1, 2, 2)
plt.plot(epochs, simple_cnn_loss, label="Simple CNN - PGD", marker='o')
plt.plot(epochs, simple_cnn_accuracy_fgsm, label="Simple CNN - FGSM", marker='s')
plt.title("Loss vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
